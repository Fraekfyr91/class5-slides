```python
def interactive_figure(alpha=0.25, p1=1, p2=2, I =10, print_sol = True):
    model = consumer.consumer(alpha=alpha, p1=p1, p2=p2, I =I)
    model.solve(print_sol=print_sol)
    model.plot()
```

```python
import ipywidgets as widgets

widgets.interact(interactive_figure,
    alpha=widgets.FloatSlider(description="$alpha$", min=0.1, max=1, step=0.05, value=0.25),
    p1=widgets.FloatSlider(description="$p_1$", min=0.1, max=5, step=0.05, value=1),
    p2=widgets.FloatSlider(description="$p_2$", min=0.1, max=5, step=0.05, value=2),
    I=widgets.FloatSlider(description="$I$", min=0.1, max=20, step=0.05, value=10),
    print_sol=widgets.Checkbox(description="print solution", value=False ),
);
```
